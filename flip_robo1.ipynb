```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_curve
from sklearn. metrics import roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
```

```python
X=df8.drop("label",axis=1)
y=df8["label"]
```

```python
X.shape,y.shape
```

```python
sc=StandardScaler()
scaled_X=sc.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test=train_test_split(scaled_X,y,test_size=0.20,random_state=0)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
lg=LogisticRegression()
```

```python
lg.fit(X_train,y_train)
```

```python
pred_lg=lg.predict(X_test)
```

```python
print(accuracy_score(y_test,pred_lg))
print('****************************************')
print(confusion_matrix(y_test,pred_lg))
print("***************************************")
print(classification_report(y_test,pred_lg))
```

```python
lg_scores=cross_val_score(lg,X,y,cv=10)
print(lg_scores)
print(lg_scores.mean(),lg_scores.std())
```

cross validation is giving same accracy so no need to consider cross validation.

```python
 y_train_pred=lg.predict_proba(X_train)
   
    
```

```python
print("lg_model train roc-auc:{}".format(roc_auc_score(y_train,y_train_pred[:,1])))
```

```python
y_test_pred=lg.predict_proba(X_test)
```

```python
print("lg_model test auc-roc:{}".format(roc_auc_score(y_test,y_test_pred[:,1])))
```

```python
y_train_pred
```

```python
fpr,tpr,thresholds=roc_curve(y_test,y_test_pred[0:,1])
```

```python
thresholds
```

```python
plt.plot(fpr,tpr,color="orange",label="ROC")
plt.plot([0,1],[0,1],"k--")
plt.xlabel("FPR")
plt.ylabel("TPR")
plt.title("ROC CURVE(LogisticRegression)")
plt.legend()
plt.show
```
